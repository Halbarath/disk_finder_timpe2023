### Functions

```python
def angular_momentum(s):
    """Calculate the angular momentum in a snapshot with pynbody."""

    J = angular_momentum_vector(s)

    return np.linalg.norm(J)


def angular_momentum_vector(s):
    """Calculate the angular momentum vector of the snapshot with pynbody."""

    m = s['mass'].in_units('kg').reshape(len(s), 1)

    v = s['vel'].in_units('m s**-1')

    p = m * v

    r = s['pos'].in_units('m')

    return np.cross(r, p).sum(axis=0)


def cart2kep(r, v, mass, central_mass):
    """
    Convert cartesian position and velocity in semi-major axis, eccentricity,
    and inclination. This saves time over calculating all Keplerian elements.

    @params
    r - (x,y,z) Cartesian Positions
    v - (vx,vy,vz) Cartesian Velocities
    mass - Particle Mass
    central_mass - Mass of central object

    @returns
    a - Semi-major axis
    ecc - Eccentricity
    inc - Inclination
    """
    
    G_SI = 6.67408e-11  # m3 kg-1 s-2

    # Gravitational Parameter
    G = G_SI
    mu = G * ( central_mass + mass )

    # Cartesian Unit Vectors
    iHat = np.array([1., 0., 0.])
    jHat = np.array([0., 1., 0.])
    kHat = np.array([0., 0., 1.])

    # Eccentricity
    h = np.cross(r, v)
    evec = 1. / mu * np.cross(v, h) - r / np.linalg.norm(r)
    ecc = np.linalg.norm(evec)

    # Semi Major Axis
    a = np.dot(h,h) / ( mu * ( 1. - ecc**2. ))

    # Inclination
    inc = np.arccos(np.dot(kHat, h) / np.linalg.norm(h))
    
    # Longitude of the Ascending Node
    n = np.cross(kHat, h)
    if inc == 0.0:
        Omega = 0.0
    else:
        Omega = np.arccos(np.dot(iHat, n) / np.linalg.norm(n))
        if np.dot(n, jHat) < 0:
            Omega = 2. * np.pi - Omega

    # Argument of Perigee
    # For Zero Inclination, Fall Back to 2D Case
    # http://en.wikipedia.org/wiki/Argument_of_periapsis
    if inc == 0.0:
        omega = np.arctan2(evec[1]/ecc,evec[0]/ecc)
    else:
        omega = np.arccos(np.dot(n, evec) / (np.linalg.norm(n) * ecc))
        if np.dot(evec, kHat) < 0:
            omega = 2. * np.pi - omega

    return a, ecc, inc, Omega, omega


def center_snapshot(s1, s2, rho_perc=90.0):
    """Center the pynbody snapshot on the center of mass of the most massive
       fragment.
    
    Calculates the center of mass for all particles with a density above the
    'rho_cut' threshold, then centers the snapshot on the center of mass. The
    threshold is determined by the percentile 'rho_perc', which should isolate
    only the dense core of the main fragment. Note that this assumes a merger
    and allows for relatively small fragments, but will not work in the case of
    two large post-impact fragments (e.g., a hit-and-run).

    Args:
        s1 (pynbody): Snapshot to be centered.
        s2 (pynbody): Snapshot used to calculate the center of mass.
        rho_perc (float): Calculate center of mass using only particles with
                          densities above this percentile.

    Returns:
        pynbody: Centered snapshot.
        
    """
    
    # Determine threshold density from percentile
    rho_cut = np.percentile(s2['rho'].in_units('g cm**-3'), rho_perc)

    s3 = s2.load_copy()[s2['rho'].in_units('g cm**-3') > rho_cut]

    xcm = np.mean(np.array(s3['pos'])[:,0])
    ycm = np.mean(np.array(s3['pos'])[:,1])
    zcm = np.mean(np.array(s3['pos'])[:,2])
    
    vxcm = np.mean(np.array(s3['vel'])[:,0])
    vycm = np.mean(np.array(s3['vel'])[:,1])
    vzcm = np.mean(np.array(s3['vel'])[:,2])
    
    del s3

    # Fix positions
    s1['pos'][:,0] -= xcm
    s1['pos'][:,1] -= ycm
    s1['pos'][:,2] -= zcm
    # Fix velocities
    s1['vel'][:,0] -= vxcm
    s1['vel'][:,1] -= vycm
    s1['vel'][:,2] -= vzcm

    # Return modified snapshot
    return s1


def determine_radius(x, y, sigmav, nbins=250, xmin=0, xmax=10, rmin=0.1):
    """Radially bins the deviation from solid body rotation to find a radius for the planet.
    
    Args:
        x (1d array): 1D array containing the radius (along x-y plane) of each particle.
        y (1d array): 1D array containing the deviation from solid body rotation of each particle.
        sigmav (float): Maximum absolute deviation from solid body rotation.
        nbins (int):  Number of bins within range (xmin,xmax).
        xmin (float): Minimum of bin range.
        xmax (float): Maximum of bin range.
        rmin (float): Minimum radius when checking deviation, to avoid noise near center.
        
    Returns:
        mids (1d array):  Midpoints of bins.
        means (1d array): Mean value of deltav in each bin.
        R_est (float):    Estimated radius of planet.
    """
    
    bins = np.linspace(xmin, xmax, num=nbins)
    
    count = binned_statistic(x, y, 'count', bins=bins)[0]
    
    stats = binned_statistic(x, y, 'median', bins=bins)
    
    means = stats[0]
    edges = stats[1]

    mids = edges[:-1] + (edges[1]-edges[0])/2
    
    
    R_est = 0

    for r, n, val in zip(mids, count, means):

        if r < rmin:
            
            R_est = r
        
        elif np.isnan(val) or (abs(val) > sigmav):
            
            R_est = r
            
            return mids, means, R_est
        
    
    return mids, means, np.nan


def orbital_velocity(M, r):
    """Calculate the Keplerian orbital velocity at r.
    
    Args:
        M (float): Central mass in kg.
        r (float): Orbital distance in m.
        
    Returns:
        float: Orbital velocity in km/s.
    """
    G_SI = 6.67408e-11  # m3 kg-1 s-2
    
    return 1e-3 * np.sqrt(G_SI * M / r)


def rotation_period(omega):
    """Convert rotation rate (Hz) to rotation period (hours)."""
    if omega > 1e-12:
        return 2 * np.pi / (3600 * float(omega))
    else:
        return 0
```

## Disk finding algorithm

### Initial guess for planet mass and radius

The purpose of this initial guess is not to accurately recover the mass and radius, but rather to provide a conservative estimate of the radius. We can then exclude the particles interior to this radius in the analysis that follows. This will significantly reduce the computational cost in the steps that follow.

```python
s = pynbody.load(stdfile)

# Total mass
M_tot = np.sum(s['mass'].in_units('5.9722e24 kg'))

rho_crit = np.percentile(s['rho'].in_units('g cm**-3'), rho_perc)


# Calculate the mass of all particles with densities above the median
s1 = s.load_copy()[s['rho'].in_units('g cm**-3') > rho_crit]

M_est_init = np.sum(s1['mass'].in_units('5.9722e24 kg'))

del s, s1


# Calculate an initial guess for the planet radius using the median density
R_est_init = (3 * M_est_init * MEARTH_G / (4 * np.pi * rho_crit))**(1/3) / REARTH_CM


print(f"Cutoff density of all particles: {rho_crit:.4f} g/cm3")
print(f"\nM_est: {M_est_init:.4f} Earth masses")
print(f"R_est: {R_est_init:.4f} Earth radii")
```

### Sliding window function for solid-body rotation

```python
def sliding_window(stdfile, d_ftvd, R_est_init, wsize=0.1, nbins=250, xmin=0, rmin=0.1):
    """Radially bins the deviation from solid body rotation to find a radius for the planet.
    
    Args:
        stdfile (str): Path to post-impact Tipsy file.
        sigmav (float): Maximum absolute deviation from solid body rotation.
        R_est_init (float): Initial naive guess for planet radius (Earth radii).
        wsize (float): Sliding window size in Earth radii.
        nbins (int):  Number of bins within range [xmin,xmax].
        xmin (float): Minimum of bin range in Earth radii.
        rmin (float): Minimum radius when checking deviation (to avoid noise near center) in Earth radii.
        
    Returns:
        mids (1d array):  Midpoints of bins.
        means (1d array): Mean value of deltav in each bin.
        R_est (float):    Estimated radius of planet.
    """

    # Load snapshot and restrict to area around central planet
    s = pynbody.load(stdfile)

    s = center_snapshot(s, s)
    
    xmax = 5 * R_est_init

    s = s[s['rxy'] < xmax]
    
    
    # Generate bins from 0 to <xmax> in Earth radii.
    bins = np.linspace(xmin, xmax, num=nbins)  

    # Get midpoints of bins
    mids = []
    for i in range(1, len(bins)):
        mids.append((bins[i]+bins[i-1])/2)

        
    # Assign particles to bins    
    s['rbin'] = np.array(pd.cut(x=s['rxy'], bins=bins, labels=mids, include_lowest=True))

    
    rbin = []
    om_r = []
    om_w = []
    om_d = []
    vt_w = []
    vt_d = []
    
    vt_exp = {}
    
    R_est = np.nan
    R_lock = 0
    
    for r in sorted(mids):
        
        # Get slice corresponding to current bin
        sbin = s[s['rbin'] == r]
        
        nbin = len(sbin)
        
        if len(sbin) == 0:
            print("Bin is empty!")
            break
            
        if r > 3:
            print("Debug stop")
            break
            
            
        rbin.append(r)
        
        # Median rotation rate in current bin (Hz)
        omega_r = np.median(sbin['vt'].in_units('km s**-1') / sbin['rxy'].in_units('km'))
        
        om_r.append(omega_r)
        
        
        # Start of sliding window
        wmin = np.max([0, r-wsize])
        
        wmid = (wmin + r) / 2
        
        
        # Get slice corresponding to sliding window
        swin = s[(s['rxy'] > wmin) & (s['rxy'] <= r)]
        
        nwin = len(swin)
        
        # Median rotation rate in sliding window (Hz)
        omega_w = np.median(swin['vt'].in_units('km s**-1') / swin['rxy'].in_units('km'))
        
        om_w.append(omega_w)
        
        # Delete up window slice
        del swin
        
        
        fd_omega = (omega_r - omega_w) / omega_w
        
        om_d.append(fd_omega)
        
        
        sbin['vt_exp'] = sbin['rxy'].in_units('km') * omega_w
        
        vt = np.median(sbin['vt_exp'])
        
        vt_w.append(vt)
        
        vt_exp[r] = vt
        
        # Fractional transverse velocity deviation
        sbin['ftvd'] = (sbin['vt'].in_units('km s**-1') - sbin['vt_exp']) / sbin['vt_exp']
        
        ftvd = np.median(sbin['ftvd'])
        
        if r > R_est_init:
            if abs(ftvd) > d_ftvd:
                if R_lock == 0:
                    R_est = r
                    R_lock = 1
        
        vt_d.append(ftvd)
        
        #sbin['deltav'] = (sbin['vt'].in_units('km s**-1') - sbin['rxy'].in_units('km') * omega_w) / (sbin['rxy'].in_units('km') * omega_w)
    

        # Delete bin slice
        del sbin
    
        
    del s
        
    return rbin, om_r, om_w, om_d, vt_w, vt_d, vt_exp, R_est
```

```python
d_ftvd = 0.05

rmids, om_r, om_w, om_d, vt_w, vt_d, vt_exp, R_est = sliding_window(stdfile, d_ftvd, R_est_init, wsize=0.3, nbins=250, xmin=0, rmin=0.1)
```

### Differentiate disk from ejecta

```python
idx_p    = []
idx_disk = []
idx_ej   = []


coll = pynbody.load(stdfile)

coll = center_snapshot(coll, coll)


central_mass = M_p * MEARTH_KG

ct = 0

for idx in range(0, len(coll)):
    
    r_i   = coll['r'][idx]
    rxy_i = coll['r'][idx]
    
    if r_i < R_est:
        idx_p.append(idx)
        
    else:
    
        mass = coll['mass'].in_units('kg')[idx]

        pos = np.array(coll['pos'].in_units('m')[idx])

        vel = np.array(coll['vel'].in_units('m s**-1')[idx])

        
        # Compute orbits
        a, e, i, Omega, omega = cart2kep(pos, vel, mass, central_mass)

        a /= 6.3710084e6  # m -> Earth radii
    
    
        # Calculate periastron distance
        r_peri = a * (1.0 - e)
        
        
        if e >= 1.0:
            idx_ej.append(idx)
        
        elif r_peri <= R_est:
            idx_p.append(idx)
            ct += 1
            
        else:
            idx_disk.append(idx)
    

if len(idx_p) > 0:
    planet = coll[idx_p]
else:
    planet = None

if len(idx_disk) > 0:
    disk = coll[idx_disk]
else:
    disk = None
    
if len(idx_ej) > 0:
    ejecta = coll[idx_ej]
else:
    ejecta = None

del coll
```

### Fluid Roche limit

```python
rho_p = np.median(planet['rho'].in_units('g cm**-3'))

rho_sat = 1.0  # g/cm3

R_roche = 2.44 * R_est * (rho_p / rho_sat)**(1/3)

print(f"Fluid Roche limit = {R_roche:.6f} Earth radii")
```
